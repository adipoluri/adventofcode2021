```python
## Load Data ##
data,ins = open('input.txt').read().split("\n\n")
reactions = [[pair for pair in swap.split(" -> ")] for swap in ins.rstrip().split("\n")]
data = data.replace("\n","")
```

```python
### Part 1 ###



def initPolymer(bonds):
    for i in range(len(data)-1):
        bond = data[i]+data[i+1]
        if bond not in bonds:
            bonds[bond] = 0
        bonds[bond] += 1
    return bonds

def countChemicals(bonds):
    chemicals = {}
    for bond in bonds:
        if bond[0] in chemicals:
            chemicals[bond[0]] += bonds[bond]
        else:
            chemicals[bond[0]] = bonds[bond]
    chemicals[data[len(data)-1]] += 1
    return chemicals
        
    
def polymerize(steps):
    bonds = {}
    bonds = initPolymer(bonds)
    
    newBonds = {}
    for i in range(steps):
        newBonds = bonds.copy()
        for reaction in reactions:
            changingBond = reaction[0]
            result1 = changingBond[0] + reaction[1] 
            result2 = reaction[1] + changingBond[1]
            
            if changingBond in newBonds:
                total = newBonds[changingBond]
                
                if result1 in bonds:
                    bonds[result1] += total
                else:
                    bonds[result1] = total
                    
                if result2 in bonds:
                    bonds[result2] += total
                else:
                    bonds[result2] = total
                   
                bonds[changingBond] -= total
            
    return bonds


polymer = polymerize(10)
countChemicals(polymer)
```
