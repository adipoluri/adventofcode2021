```python
### Part 1 ###
openVals = ["[","{","<","("]
closeVals = ["]","}",">",")"]
scores = {')':3,']':57,'}':1197,'>':25137}

'''
): 3 points.
]: 57 points.
}: 1197 points.
>: 25137 points.

'''

def equivalentBracket(left,right):
    return openVals.index(left) == closeVals.index(right)
    

def findCompilerScore():
    score = 0
    for line in data:
        stack = []
        for bracket in line:
            if bracket in openVals:
                stack.append(bracket)
            else:
                topVal = stack.pop()
                if not equivalentBracket(topVal,bracket):
                    score += scores[bracket]
    return score

        
print(findCompilerScore())
```

```python
### Part 2 ###
openVals = ["[","{","<","("]
closeVals = ["]","}",">",")"]
scores = {')':1,']':2,'}':3,'>':4}

'''
): 3 points.
]: 57 points.
}: 1197 points.
>: 25137 points.

'''

def equivalentBracket(left,right):
    return openVals.index(left) == closeVals.index(right)

def getEquivBracket(left):
    return closeVals[openVals.index(left)]


def findCompilerScore():
    score = []
    for line in data:
        stack = []
        for bracket in line:
            if bracket in openVals:
                stack.append(bracket)
            else:
                topVal = stack.pop()
                if not equivalentBracket(topVal,bracket):
                    continue
        if (len(stack)) != 0:
            currScore = 0
            for bracket in range(len(stack)):
                currScore *= 5
                currScore += scores[getEquivBracket(stack.pop())]
            score.append(currScore)
        
    return sorted(score)[len(score)//2-1]

        
print(findCompilerScore())
```
