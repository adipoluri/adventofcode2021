```python
### Part 1 ###


def formatData():
    coords = []
    for line in data:
        entry = line.replace(" ","").replace("\n","").replace("->",",").split(",")
        for i in range(len(entry)):
            entry[i] = int(entry[i])
        coords.append(entry)
        
    return coords

def mapVents():
    coords = formatData()

    nx = max(max([(c[0], c[2])for c in coords])) + 1
    ny = max(max([(c[1], c[3])for c in coords])) + 1

    board = [[0 for i in range(nx)] for j in range(ny)]

    for x1, y1, x2, y2 in coords:
        xmin, dx = min(x1, x2), abs(x2 - x1)
        ymin, dy = min(y1, y2), abs(y2 - y1)
        if x1 == x2:
            for i in range(dy + 1):
                board[i + ymin][x1] += 1
        elif y1 == y2:
            for i in range(dx + 1):
                board[y1][i + xmin] += 1


    return sum([1 for line in board for num in line if num > 1])

print(mapVents())

```

```python
### Part 2 ###


def mapVents():
    coords = formatData()

    nx = max(max([(c[0], c[2])for c in coords])) + 1
    ny = max(max([(c[1], c[3])for c in coords])) + 1

    board = [[0 for i in range(nx)] for j in range(ny)]

    for x1, y1, x2, y2 in coords:
        xmin, dx = min(x1, x2), abs(x2 - x1)
        ymin, dy = min(y1, y2), abs(y2 - y1)
        if x1 == x2:
            for i in range(dy + 1):
                board[i + ymin][x1] += 1
        elif y1 == y2:
            for i in range(dx + 1):
                board[y1][i + xmin] += 1
        else:
            for i in range(max(dx,dy)+1):
                xdir = 1 if x1 < x2 else -1
                ydir = 1 if y1 < y2 else -1
                board[ydir*i + y1][xdir*i + x1] += 1

    return sum([1 for line in board for num in line if num > 1])

print(mapVents())

```

```python
### Visualisation ###
import sys, pygame
from pygame import rect, Rect

data = open('day5.txt').readlines()

#dimensions
size = width, height = 1000,1000

#colors
white = 255,255,255
black = 0,0,0
grey = 112, 112, 112
gold = 148, 124, 0
yellow = 255, 214, 0
green = 38, 189, 0
red =255, 0, 0

#constants
tableSize = 5
startY = 20
startX = 15
moveX = 100
moveRowY = 100
startDataX = 1020
counterStart = 1075



### ------------ Solving Functions ------------ ### 
def formatData():
    coords = []
    for line in data:
        entry = line.replace(" ","").replace("\n","").replace("->",",").split(",")
        for i in range(len(entry)):
            entry[i] = int(entry[i])
        coords.append(entry)
        
    return coords

def mapVents():
    board = [[0 for i in range(1000)] for j in range(1000)]

    return(board)

def updateVents(counter, board, coords):
    for i in range(5):
        x1, y1, x2, y2 = coords[counter+i]

        xmin, dx = min(x1, x2), abs(x2 - x1)
        ymin, dy = min(y1, y2), abs(y2 - y1)
        if x1 == x2:
            for i in range(dy + 1):
                board[i + ymin][x1] += 1
        elif y1 == y2:
            for i in range(dx + 1):
                board[y1][i + xmin] += 1
        else:
            for i in range(max(dx,dy)+1):
                xdir = 1 if x1 < x2 else -1
                ydir = 1 if y1 < y2 else -1
                board[ydir*i + y1][xdir*i + x1] += 1
    
    return(board)
            

### ------------ End Solving Functions ------------ ### 

    
def drawArt(screen,ind):
    font = pygame.font.Font('Minecraft.ttf', 20)
    if(ind % 2 == 0):
        text = font.render("Vent OS|", True, white, black)
    else:
        text = font.render("Vent OS", True, white, black)
    textRect = text.get_rect()
    textRect.center = (width/2, height/2)
    screen.blit(text, textRect)
                       
            
    
def drawVents(screen,data):
    x = startX
    y = startY
    for y in range(len(data)):
        for x in range(len(data[0])):
            if(data[y][x] >= 2):
                pygame.draw.rect(screen, red,(x,y,1,1))
            elif(data[y][x] == 1):
                pygame.draw.rect(screen, grey,(x,y,1,1))
    

        
def visualisation():
    pygame.init()
    screen = pygame.display.set_mode(size)
    pygame.display.set_caption('Day5 Visualiser')
    
    coords = formatData()
    board = mapVents()
    
    updateSpeed = 500
    counter = -6
    
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT: 
                pygame.quit()
                return
        
        if counter < 0: #Starting Screen
            screen.fill(black)
            drawArt(screen,counter)
            counter += 1
            
        elif counter < 500:  #VIsualization
            updateSpeed = 1
            
            board = updateVents(counter,board,coords)
            counter += 5

            screen.fill(black)
            drawVents(screen,board)  
            
        pygame.display.update()
        pygame.time.wait(updateSpeed)
        
        

            
        
        
visualisation()
```
