```python
### Part 1 ###

xBound = len(data[0])
yBound = len(data)

def isBasin(x,y,data):
    lookup = [[1,0],[-1,0],[0,1],[0,-1]]
    
    for i in range(4):
        xLookup = lookup[i][0]
        yLookup = lookup[i][1]
        if (x+xLookup>=xBound) or (x+xLookup<0) or (y+yLookup>=yBound) or (y+yLookup<0):
            continue
            
        if(data[y+yLookup][x+xLookup]<=data[y][x]):
            return False
        
    return True
        
    
def findBasins():
    riskLevel = 0
    for y in range(len(data)):
        for x in range(len(data[0])):
            if(isBasin(x,y,data)):
                riskLevel += 1 + data[y][x]

    return riskLevel

print(findBasins())
        
```

```python
### Part 2 ###

xBound = len(data[0])
yBound = len(data)


def isBasin(x,y,data):
    lookup = [[1,0],[-1,0],[0,1],[0,-1]]
    
    for i in range(4):
        xLookup = lookup[i][0]
        yLookup = lookup[i][1]
        if (x+xLookup>=xBound) or (x+xLookup<0) or (y+yLookup>=yBound) or (y+yLookup<0):
            continue
            
        if(data[y+yLookup][x+xLookup]<=data[y][x]):
            return False
        
    return True
        
    
def findBasinCenters():
    basins = []
    for y in range(len(data)):
        for x in range(len(data[0])):
            if(isBasin(x,y,data)):
                basins.append([x,y])
    return basins


def valid(x,y,v):
    return (x<xBound) and (x>=0) and (y<yBound) and (y>=0) and data[y][x]!=9 and v[y][x] == 0

    
def dfs(x,y,v):
    
    v[y][x] = 1
    
    if(valid(x+1,y,v)):
        dfs(x+1,y,v)
    if(valid(x-1,y,v)):
        dfs(x-1,y,v)
    if(valid(x,y-1,v)):
        dfs(x,y-1,v)
    if(valid(x,y+1,v)):
        dfs(x,y+1,v)


def findBasinsSize(basinCenters):
    basinSizes = []
    counter = 0
    for basins in basinCenters:
        visited = [[0 for x in range(100)] for y in range(100)]
        dfs(basins[0],basins[1],visited)
        basinSizes.append(sum([sum(line) for line in visited]))
    return basinSizes
        
    
bs= sorted(findBasinsSize(findBasinCenters()))

print(str(bs[len(bs)-1]*bs[len(bs)-2]*bs[len(bs)-3]))
        
```

```python
### Visualisation ###

import sys, pygame
from pygame import rect, Rect
from random import random,uniform
import math

data = [[int(digit) for digit in list(num.replace("\n",""))] for num in open('day9.txt').readlines()]

#dimensions
size = width, height = 1000,1000

#nord colors
b0 = pygame.Color("#2E3440")
b1 = pygame.Color("#3B4252")
b2 = pygame.Color("#434C5E")
b3 = pygame.Color("#4C566A")

w0 = pygame.Color("#D8DEE9")
w1 = pygame.Color("#E5E9F0")
w2 = pygame.Color("#ECEFF4")

blue0 = pygame.Color("#8FBCBB")
blue1 = pygame.Color("#88C0D0")
blue2 = pygame.Color("#81A1C1")
blue3 = pygame.Color("#5E81AC")

red = pygame.Color("#BF616A")
orange = pygame.Color("#D08770")
yellow = pygame.Color("#EBCB8B")
green = pygame.Color("#A3BE8C")
purple = pygame.Color("#B48EAD")


#constants
startY = 50
startX = 50

xBound = len(data[0])
yBound = len(data)




### ------------ Solving Functions ------------ ### 

def isBasin(x,y,data):
    lookup = [[1,0],[-1,0],[0,1],[0,-1]]
    
    for i in range(4):
        xLookup = lookup[i][0]
        yLookup = lookup[i][1]
        if (x+xLookup>=xBound) or (x+xLookup<0) or (y+yLookup>=yBound) or (y+yLookup<0):
            continue
            
        if(data[y+yLookup][x+xLookup]<=data[y][x]):
            return False
        
    return True
        
    
def findBasinCenters(data):
    basins = []
    for y in range(len(data)):
        for x in range(len(data[0])):
            if(isBasin(x,y,data)):
                basins.append([x,y])
    return basins

def valid(x,y,v):
    return (x<xBound) and (x>=0) and (y<yBound) and (y>=0) and data[y][x]!=9 and v[y][x] == 0

    
def dfs(x,y,v,ret):
    
    v[y][x] = 1
    ret.append((x,y))
    
    if(valid(x,y+1,v)):
        dfs(x,y+1,v,ret)
    if(valid(x,y-1,v)):
        dfs(x,y-1,v,ret)
    if(valid(x+1,y,v)):
        dfs(x+1,y,v,ret)
    if(valid(x-1,y,v)):
        dfs(x-1,y,v,ret)



### ------------ End Solving Functions ------------ ### 


    
def sonarArt(screen,radius):
    xCoord = 0
    yCoord = 0
    for y in range(len(data)):
        for x in range(len(data[0])):
            pythag = (xCoord+5-width/2)*(xCoord+5-width/2) + (yCoord+5-height/2)*(yCoord+5-height/2)
            if(radius*radius > pythag):
                pygame.draw.rect(screen, depthColour(data[y][x]),(xCoord,yCoord,10,10))
            xCoord += 10
        yCoord += 10
        xCoord = 0
    pygame.draw.circle(screen, w1,(width/2,height/2),radius+5,10)


    

def drawArt(screen,ind):
    font = pygame.font.Font('Minecraft.ttf', 20)
    if(ind % 2 == 0):
        text = font.render("Smoke OS|", True, w0, b0)
    else:
        text = font.render("Smoke OS", True, w0, b0)
    textRect = text.get_rect()
    textRect.center = (width/2, height/2)
    screen.blit(text, textRect)
                    
        
def depthColour(depth):
    st = 20
    colors = [(st, 255-st, 255-st),
              (st, 200-st, 255-st),
              (st, 150-st, 255-st),
              (st, 100-st, 255-st),
              (st, 50-st, 255-st),
              (st, st, 255-st),
              (50-st, st, 255-st),
              (100-st, st, 255-st),
              (150-st, st, 255-st),
              (200-st, st, 255-st)]
    
    return colors[9-depth]


def drawGrid(screen,data):
    xCoord = 0
    yCoord = 0
    for y in range(len(data)):
        for x in range(len(data[0])):
            pygame.draw.rect(screen, depthColour(data[y][x]),(xCoord,yCoord,10,10))
            xCoord += 10
        yCoord += 10
        xCoord = 0

def drawWells(screen,data,counter):
    for y in range(counter-1000):
        pygame.draw.rect(screen, b0,(data[y][0]*10,data[y][1]*10,10,10))

        
def visualisation():
    pygame.init()
    screen = pygame.display.set_mode(size)
    pygame.display.set_caption('Day8 Visualiser')
    
    
    basinCenters = findBasinCenters(data)
    ret = []
    for basins in basinCenters:
        visited = [[0 for x in range(100)] for y in range(100)]
        dfs(basins[0],basins[1],visited,ret)

    updateSpeed = 500
    counter = -6
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT: 
                pygame.quit()
                return
        
        screen.fill(b0)
        if counter < 0: #Starting Screen
            drawArt(screen,counter)
            
        elif counter < 1000:
            updateSpeed = 1
            sonarArt(screen,counter)
            
        elif counter > 1000:  #Visualization
            updateSpeed = 1
            drawGrid(screen,data)
            drawWells(screen,ret,counter)
            
        if updateSpeed != 1:
            pygame.time.wait(updateSpeed)
        
        counter += 1
        pygame.display.update()
        

        
visualisation()
```
