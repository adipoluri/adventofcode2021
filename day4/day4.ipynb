--- Day 4: Giant Squid ---
You're already almost 1.5km (almost a mile) below the surface of the ocean, already so deep that you can't see any sunlight. What you can see, however, is a giant squid that has attached itself to the outside of your submarine.

Maybe it wants to play bingo?

Bingo is played on a set of boards each consisting of a 5x5 grid of numbers. Numbers are chosen at random, and the chosen number is marked on all boards on which it appears. (Numbers may not appear on all boards.) If all numbers in any row or any column of a board are marked, that board wins. (Diagonals don't count.)

At this point, the third board wins because it has at least one complete row or column of marked numbers (in this case, the entire top row is marked: 14 21 17 24 4).

The score of the winning board can now be calculated. Start by finding the sum of all unmarked numbers on that board; in this case, the sum is 188. Then, multiply that sum by the number that was just called when the board won, 24, to get the final score, 188 * 24 = 4512.

To guarantee victory against the giant squid, figure out which board will win first. What will your final score be if you choose that board?

--- Part Two ---
On the other hand, it might be wise to try a different strategy: let the giant squid win.

You aren't sure how many bingo boards a giant squid could play at once, so rather than waste time counting its arms, the safe thing to do is to figure out which board will win last and choose that one. That way, no matter which boards it picks, it will win for sure.

In the above example, the second board is the last to win, which happens after 13 is eventually called and its middle column is completely marked. If you were to keep playing until this point, the second board would have a sum of unmarked numbers equal to 148 for a final score of 148 * 13 = 1924.

Figure out which board will win last. Once it wins, what would its final score be?

Your puzzle answer was 19012

```python
### Part 1 & Part 2###

tableSize = 5

def getRow(string):
    return string.strip().replace("  ", " ").replace("\n","").split(" ")


def tablize(data):
    tbls = []
    
    for x in range(2,len(data),6):
        tbl = []
        for y in range(tableSize):
            tbl.append(getRow(data[x+y]))
        tbls.append(tbl)
        
    return tbls


def rollNumber(tables,number):
    for i in range(len(tables)):
        for x in range(tableSize):
            for y in range(tableSize):
                if(tables[i][x][y]==number):
                    tables[i][x][y] = "X"
                    
                    
def checkRows(table):
    for i in range(tableSize):
        if "".join(table[i]) == "XXXXX":
            return True
    return False

def checkColumns(table):
    for x in range(tableSize):
        column = ""
        for y in range(tableSize):
            column += table[y][x]
        if column == "XXXXX":
            return True
    return False

     
def getSum(table):
    sum = 0
    for x in range(tableSize):
        for y in range(tableSize):
            if table[x][y] != "X":
                sum += int(table[x][y])
    return sum
             
    
def checkBingo(tables):
    retVal = []
    for i in range(len(tables)):
        if checkRows(tables[i]):
            retVal.append(i)
            continue
        if checkColumns(tables[i]):
            retVal.append(i)
    return retVal      
    
    
def bingo(data):
    rolls = data[0][0:len(data[0])-1].split(",")
    tables = tablize(data)
    
    res = []
    score = []
    for roll in rolls:
        rollNumber(tables,roll)
        bingo = checkBingo(tables)
        if(len(bingo) != 0):
            for i in range(len(bingo)):
                score.append(getSum(tables[bingo[i]-i])*int(roll))
                res.append(tables.pop(bingo[i]-i))
    print(score[0])
    print(score[len(score)-1])

    
bingo(data)

```

```python
## Visualization ###

import sys, pygame
from pygame import rect, Rect

data = open('day4.txt').readlines()

#dimensions
size = width, height = 1325,1015

#colors
white = 255,255,255
black = 0,0,0
grey = 112, 112, 112
gold = 148, 124, 0
yellow = 255, 214, 0
green = 38, 189, 0

#constants
tableSize = 5
startY = 20
startX = 15
moveX = 100
moveRowY = 100
startDataX = 1020
counterStart = 1075



### ------------ Solving Functions ------------ ### 
def getRow(string):
    return string.strip().replace("  ", " ").replace("\n","").split(" ")

def tablize(data):
    tbls = []
    
    for x in range(2,len(data),6):
        tbl = []
        for y in range(tableSize):
            tbl.append(getRow(data[x+y]))
        tbls.append(tbl)
        
    return tbls

def rollNumber(tables,number,winners):
    for i in range(len(tables)):
        if i in winners:
            continue
        for x in range(tableSize):
            for y in range(tableSize):
                if(tables[i][x][y]==number):
                    tables[i][x][y] = "X"

def checkRows(table):
    for i in range(tableSize):
        if "".join(table[i]) == "XXXXX":
            return True
    return False

def checkColumns(table):
    for x in range(tableSize):
        column = ""
        for y in range(tableSize):
            column += table[y][x]
        if column == "XXXXX":
            return True
    return False

     
def getSum(table):
    sum = 0
    for x in range(tableSize):
        for y in range(tableSize):
            if table[x][y] != "X":
                sum += int(table[x][y])
    return sum
             
    
def checkBingo(tables,winners):
    retVal = []
    for i in range(len(tables)):
        if i in winners:
            continue
        if checkRows(tables[i]):
            retVal.append(i)
            continue
        if checkColumns(tables[i]):
            retVal.append(i)
    return retVal      

### ------------ End Solving Functions ------------ ### 


def getGrey(index):
    i = index+1
    return(255/i,255/i,255/i)

def drawRollText(screen,rolls, rollCounter):
    font = pygame.font.Font('Minecraft.ttf', 20)
    
    pygame.draw.rect(screen, white,(counterStart,15,30,30))
    pygame.draw.rect(screen, black,(counterStart+2,17,26,26))
    
    if rollCounter<len(rolls)-7:
        for i in range(7):
            text = font.render(rolls[rollCounter+i], True, getGrey(i), black)
            textRect = text.get_rect()
            textRect.center = (counterStart+i*30+15,32)
            screen.blit(text, textRect)    
    else:
        for i in range(len(rolls)-rollCounter):
            text = font.render(rolls[rollCounter+i], True, getGrey(i), black)
            textRect = text.get_rect()
            textRect.center = (counterStart+i*30+15,32)
            screen.blit(text, textRect) 

    
def drawArt(screen,ind):
    font = pygame.font.Font('Minecraft.ttf', 20)
    if(ind % 2 == 0):
        text = font.render("Squid OS|", True, white, black)
    else:
        text = font.render("Squid OS", True, white, black)
    textRect = text.get_rect()
    textRect.center = (width/2, height/2)
    screen.blit(text, textRect)
                       
            
def drawScoreText(screen,scores):
    font = pygame.font.Font('Minecraft.ttf', 12)
    x = startDataX
    y = 60
    counter = 1
    for board,score in scores:
        colorMain = white
        
        if(counter == 1):
            colorMain = green
        if(counter == 100):
            colorMain = green
        if(counter == 51):
            x += 150
            y = 60
            
        text = font.render("["+str(counter)+"]: Board "+str(board+1)+" = " + str(score), True, colorMain, black)
        textRect = text.get_rect()
        textRect.center = (x, y)
        textRect.left = x
        screen.blit(text, textRect)
        
        colorMain = white
        y += 16
        counter += 1
        
def drawBoardHeader(screen,x,y,index,winCon):
    colorMain = grey
    colorSec = grey
    if winCon:
        colorMain = gold
        colorSec = white
        
    pygame.draw.rect(screen, colorMain,(x-5,y-5,80,90))
    pygame.draw.rect(screen, black,(x-3,y-3,76,86))
    
    font = pygame.font.Font('Minecraft.ttf', 15)
    text = font.render(str(index), True, colorSec, black)
    textRect = text.get_rect()
    textRect.center = (x+35, y+82)
    screen.blit(text, textRect)

def drawBoard(screen,xstart,ystart,ind,data,winCon):
    font = pygame.font.Font('Minecraft.ttf', 10)
    x = xstart
    y = ystart
    drawBoardHeader(screen,xstart,ystart,ind+1,winCon)
    for i in range(tableSize):
        for j in range(tableSize):
            if winCon and (data[i][j] == "X"):
                text = font.render(data[i][j], True, yellow, black)
            elif data[i][j] == "X":
                text = font.render(data[i][j], True, white, black)
            elif winCon:
                text = font.render(data[i][j], True, gold, black)
            else:
                text = font.render(data[i][j], True, grey, black)
            textRect = text.get_rect()
            textRect.center = (x+i*15+5, y+j*15+5)
            screen.blit(text, textRect)

    
def drawBoards(screen,data,winners):
    x = startX
    y = startY
    counter = 0
    winCon = False
    for i in range(len(data)):
        if i in winners:
            winCon = True
        if(counter >= 10):
            x = startX
            y += moveRowY
            counter = 0
            
        drawBoard(screen,x,y,i,data[i],winCon)
        winCon = False
        counter += 1
        x += moveX
        
def visualisation():
    pygame.init()
    screen = pygame.display.set_mode(size)
    pygame.display.set_caption('Day4 Visualiser')
    
    rolls = data[0][0:len(data[0])-1].split(",")
    tables = tablize(data)
    
    updateSpeed = 500
    
    winners = []
    scores = []
    
    rollCounter = -6
    currentRoll = ""
    
    
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT: 
                pygame.quit()
                return
        
        if rollCounter < 0: #Starting Screen
            screen.fill(black)
            drawArt(screen,rollCounter)
            rollCounter += 1
            
        elif rollCounter < len(rolls):  #VIsualization
            updateSpeed = 200
            currentRoll = rolls[rollCounter]
            rollNumber(tables,currentRoll,winners)
            bingo = checkBingo(tables,winners)
            
            if(len(bingo) != 0):
                for i in range(len(bingo)):
                    scores.append([bingo[i],getSum(tables[bingo[i]])*int(currentRoll)])
                    winners.append(bingo[i])
                    
            rollCounter += 1
                    
            screen.fill(black)
            drawBoards(screen,tables,winners)  
            drawRollText(screen,rolls,rollCounter)
            drawScoreText(screen,scores)
            
        pygame.display.update()
        pygame.time.wait(updateSpeed)
        
        

            
        
        
visualisation()
```
